# Visualizing the data gathered from the PSNR/BitRate Parser and the BDrate computation function from BDrate.py (by Shengbin Meng) implemented on main.py

### Importing the relevant libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

### Loading contents from the JSON config file

```python
file_cfg = open('configs.json')
configs = json.load(file_cfg)
file_cfg.close()
```

### Initializing pandas DataFrame with the resulting information

```python
df = pd.read_csv("/home/arthurscarpatto/VC/video-coding/results.csv") 
vtmdf = df.iloc[:16, :]
vvencdf = df.iloc[16:, :]
```

### Visualizing the VTM results

```python
fig, axes = plt.subplots(2, 1, figsize=(15,5))
sns.barplot(data=vtmdf, x='v', y='psnr', hue='qsize', ax=axes[0]).set(title='VTM Test Run: Correlation Between Quantization Parameters and PSNR/BitRate', xlabel='')
sns.barplot(data=vtmdf, x='v', y='bit_rate', hue='qsize', ax=axes[1]).set(xlabel='Videos')
```

### Visualizing the vvenc results

```python
fig, axes = plt.subplots(2, 1, figsize=(15,5))
sns.barplot(data=vvencdf, x='v', y='psnr', hue='qsize', ax=axes[0]).set(title='VVenC Test Run: Correlation Between Quantization Parameters and PSNR/BitRate', xlabel='')
sns.barplot(data=vvencdf, x='v', y='bit_rate', hue='qsize', ax=axes[1]).set(xlabel='Videos')
```

### Visualizing the average difference (BD-Rate) between VTM and vvenc

```python
from BDrate import BD_RATE

# Initializing useful data structures
avg_diffs = []
labels = []

for v in configs['videos']:
    # Calculating the BD-Rate
    index_div = v.index("_")
    v_name = v[:index_div]
    videodf = df[df['v'] == v_name]
    r1 = videodf['bit_rate'].iloc[:4]
    r2 = videodf['bit_rate'].iloc[4:]
    psnr1 = videodf['psnr'].iloc[:4]
    psnr2 = videodf['psnr'].iloc[4:]
    avg_diffs.append(BD_RATE(r1, psnr1, r2, psnr2))
    
    # Getting the names of the videos, without the resolution
    div_index = v.index("_")
    labels.append(v[:div_index])

# Plotting the graph
plt.figure(figsize=(15,5))
sns.set_style('dark')
sns.lineplot(y=avg_diffs, x=labels, palette=['b']).set(ylabel='Average difference', title='BD-Rate: Average Difference Between VTM and VVenC for Four Sequences', xlabel='Sequences')
plt.show()
```

### Visualizing the difference between VTM and vvenc by means of a BitRate vs PSNR scatter plot

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 10))
fig.suptitle('PSNRxBitRate per Sequence: VTM and VVenC Compared', fontsize=16)

# first sequence
firstvtm = vtmdf[:4]
firstvvenc = vvencdf[:4]
axes[0][0].plot(firstvtm.bit_rate, firstvtm.psnr, linestyle='--', marker='x', c='b', label='VTM')
axes[0][0].plot(firstvvenc.bit_rate, firstvvenc.psnr, linestyle='--', marker='o', c='r')
axes[0][0].set_title('BasketballPass')
axes[0][0].legend(['VTM', 'vvenc'])

# second sequence
secondvtm = vtmdf[4:8]
secondvvenc = vvencdf[4:8]
axes[0][1].plot(secondvtm.bit_rate, secondvtm.psnr, linestyle='--', marker='x', c='b')
axes[0][1].plot(secondvvenc.bit_rate, secondvvenc.psnr, linestyle='--', marker='o', c='r')
axes[0][1].set_title('Cactus')
axes[0][1].legend(['VTM', 'vvenc'])

# third sequence
thirdvtm = vtmdf[8:12]
thirdvvenc = vvencdf[8:12]
axes[1][0].plot(thirdvtm.bit_rate, thirdvtm.psnr, linestyle='--', marker='x', c='b')
axes[1][0].plot(thirdvvenc.bit_rate, thirdvvenc.psnr, linestyle='--', marker='o', c='r')
axes[1][0].set_title('RaceHorsesC')
axes[1][0].legend(['VTM', 'vvenc'])

# fourth sequence
fourthvtm = vtmdf[12:16]
fourthvvenc = vvencdf[12:16]
axes[1][1].plot(fourthvtm.bit_rate, fourthvtm.psnr, linestyle='--', marker='x', c='b')
axes[1][1].plot(fourthvvenc.bit_rate, fourthvvenc.psnr, linestyle='--', marker='o', c='r')
axes[1][1].set_title('FourPeople')
axes[1][1].legend(['VTM', 'vvenc'])

plt.savefig('./graphs/PSNRxBitRate by Sequence')
```
